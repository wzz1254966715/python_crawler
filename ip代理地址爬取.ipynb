```python
import re
import json
from selenium import webdriver
```

```python
class IpCrawler(object):
    """
        爬取最新高匿ip, 保存本地
        使用时使用json解码就可以获得列表
    """
    def __init__(self, path=r'F:\爬虫头域\ip.txt'):
        """
            path: 文件保存路径, 默认'F:\爬虫头域\ip.txt'
        """
        self.url = r'https://www.xicidaili.com/nn/'
        self.path = path
        
    
    def start(self):
        """
            执行下载
        """
        # 设置不打开浏览器
        option = webdriver.ChromeOptions()
        option.set_headless()
        # 启动浏览器驱动
        driver = webdriver.Chrome(executable_path=r"C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe", options=option)
        # 请求网页
        driver.get(url)
        # 获取内容
        content = driver.find_element_by_tag_name('tbody').text
        # 格式化内容
        patten = r'(HTT.*)\n(\d)'
        repl = r'\1 \2'
        content = re.sub(patten, repl, content).split('\n')
        content.pop(0)
        l = []
        for i in content:
            k = []
            patten_ip = re.compile(r'([\d\.]+) ')
            patten_h = re.compile(r'HTTPS?')
            k.append(patten_h.search(i).group().lower())
            k.append(patten_ip.match(i).group(1))
            l.append(k)
        l = json.dumps(l)
        with open(self.path, 'w') as f:
            f.write(l)
        print('Successful!')
        
    def display(self):
        with open(self.path, 'r') as f:
            n = f.read()
        ip_list = json.loads(n)
        return ip_list
```

```python
if __name__ =='__main__':
    ip = IpCrawler()
    ip.start()
    ip_list = ip.display()
    for i in ip_list:
        print(i)
```
